```python
import pandas as pd
```

```python
print ("START")

startYear = 1986
endYear = 1987
header = 1
df = pd.DataFrame() 

for year in range (startYear, endYear ):
    print (f"START JAHR {year}")
    if year < 1987:
        end= "TMW"
    else :
        end= "3hmw"
    fileName = f"Staub/PM10_{year}_{end}.xlsx"
    dfRead = pd.read_excel (fileName,header=header)
    dfOneYear = dfRead[dfRead["Station"].str.startswith ("Nürnberg / Bahnhof", na=False)]
    dfOneYear = dfOneYear.drop ("Komponente", axis = 1)
    dfOneYear = dfOneYear.rename(columns={"TMW": "PM10"})
    dfOneYear["Station"] = "Nürnberg"
    dfOneYear["PM10"] = pd.to_numeric(dfOneYear["PM10"], errors='coerce')
    df = pd.concat ([df, dfOneYear], ignore_index=True)
    print (len (df))

print (df.head ())
df.to_csv ("Staub_1986.csv", index=False, sep=";", encoding="utf-8")

print ("ENDE")
```

```python

print ("START")

startYear = 1987
endYear = 2006
header = 1
df = pd.DataFrame() 

for year in range (startYear, endYear ):
    print (f"START JAHR {year}")
    fileName = f"Staub/PM10_{year}_3hmw.xlsx"
    dfRead = pd.read_excel (fileName,header=header)
    dfcomp = dfRead[dfRead["Station"].str.startswith ("Nürnberg / Bahnhof", na=False)]
    dfcomp = dfcomp.drop ("Komponente", axis = 1)
    
    dfOneYear = dfcomp.melt (id_vars=["Station", "Datum"], 
                     var_name="Stunde",      
                     value_name="PM10")
    dfOneYear["Station"] = "Nürnberg"

    dfOneYear['Stunde'] = dfOneYear['Stunde'].astype(str)
    
    # 2. Falls "24:00:00" vorkommt, korrigieren (optional, falls Sie 24 als 0 Uhr werten wollen)
    # dfOneYear['Stunde'] = dfOneYear['Stunde'].str.replace('24:00:00', '00:00:00', regex=False)
    
    # 3. Sicher in Pandas Timedelta konvertieren (behebt das Typ-Problem aus Excel)
    timedelta_spalte = pd.to_timedelta(dfOneYear['Stunde'], errors='coerce')
    
    # 4. Die Stunden exakt als mathematische Ganzzahl berechnen
    # (Berücksichtigt auch Tage: 1 Tag und 0 Std wird automatisch zu 24 Std)
    dfOneYear['Stunde'] = timedelta_spalte.dt.components['hours']
    dfOneYear['PM10'] = pd.to_numeric(dfOneYear['PM10'], errors='coerce')
    
    df = pd.concat ([df, dfOneYear], ignore_index=True)
    print ((df))

#print (df.head ())
df.to_csv ("Staub_1987.csv", index=False, sep=";", encoding="utf-8")

print ("ENDE")
```

```python

print ("START")

startYear = 2005
endYear = 2025
header = 1
df = pd.DataFrame() 

for year in range (startYear, endYear ):
    print (f"START JAHR {year}")
    add = ""
    if year <= 2011:
        add = "/Bahnhof"
    fileName = f"Staub/PM10_{year}.xlsx"
    dfRead = pd.read_excel (fileName,header=header)
    dfColumns = [c for c in dfRead.columns if str(c).startswith(f'Nürnberg{add}')]
    dfColumn = dfColumns[0] 
    dfOneYear = dfRead [["Zeitpunkt",dfColumn]]
    dfOneYear = dfOneYear [dfOneYear["Zeitpunkt"] != "0/0"]    
    dfOneYear = dfOneYear.rename (columns= {dfColumn: "PM10"})
    dfOneYear[["Datum", "Stunde"]] = dfOneYear["Zeitpunkt"].str.split(' ', expand=True)
    dfOneYear["Station"] = "Nürnberg"
    rang = ["Station", "Datum", "Stunde", "PM10"]
    dfOneYear['Stunde'] = dfOneYear['Stunde'].str.replace('24:', '00:', regex=False)
    dfOneYear['Stunde'] = pd.to_datetime(dfOneYear['Stunde'].str.strip(), format='%H:%M').dt.hour
    dfOneYear['PM10'] = pd.to_numeric(dfOneYear['PM10'], errors='coerce')
    dfOneYear = dfOneYear [rang]
    df = pd.concat ([df, dfOneYear], ignore_index=True)

print (df.head ())
df.to_csv ("Staub_2005.csv", index=False, sep=";", encoding="utf-8")

print ("ENDE")
```
