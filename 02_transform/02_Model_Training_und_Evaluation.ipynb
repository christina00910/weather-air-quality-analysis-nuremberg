# 📓 Modul 2: Modell-Training & Evaluation
**Ziel dieses Notebooks:** Den aufbereiteten Datensatz nutzen, um verschiedene Machine-Learning-Algorithmen zu testen. Wir identifizieren das Modell, das Luftqualität (z. B. PM10 oder NO2) basierend auf Wetter- und Vergangenheitsdaten am besten vorhersagt.

## 📋 To-Do / Berechnungen:
- [ ] **Daten-Import:** Einlesen der `train_ready_data.csv`.
- [ ] **Definition von Features (X) und Target (y):** - Was wollen wir vorhersagen? (z.B. `y = PM10`). Was sind die Prädiktoren? (`X = Temperatur, Wind, Monat, Stunde, ...`).
- [ ] **Chronologischer Train/Test-Split ⚠️ WICHTIG:**
    - **KEIN** zufälliger Split (kein `train_test_split` von sklearn)!
    - Split-Beispiel: Train-Set = 1980 bis 2020, Test-Set = 2021 bis 2024. Wir dürfen nicht mit der Zukunft die Vergangenheit trainieren (Data Leakage).
- [ ] **Baseline-Modell trainieren:**
    - Ein einfaches Modell (z.B. Lineare Regression) aufsetzen, um einen Vergleichswert zu haben.
- [ ] **Komplexes Modell trainieren:**
    - Z.B. `RandomForestRegressor` oder `XGBoost`.
- [ ] **Modell-Evaluation (Metriken berechnen):**
    - Berechnen von **RMSE** (Root Mean Squared Error), **MAE** (Mean Absolute Error) und **R²** auf dem Test-Set.
- [ ] **Feature Importance Plotten:**
    - Welcher Faktor beeinflusst die Luftqualität am meisten? (Wind? Temperatur? Die Uhrzeit?).
