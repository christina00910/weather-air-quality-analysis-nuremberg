# 📓 Modul 1: EDA & Feature Engineering
**Ziel dieses Notebooks:** Die stündlichen Rohdaten in einen "maschinell lernbaren" Zustand zu überführen. KI-Modelle können mit reinen Zeitstempeln wenig anfangen, sie brauchen Muster. Hier extrahieren wir diese Muster und bauen den finalen Datensatz für das Training.

## 📋 To-Do / Berechnungen:
- [ ] **Daten-Import:** Einlesen der `Schadstoff_Wetter.csv`.
- [ ] **Umgang mit fehlenden Werten (NaNs):** - Wie füllen wir Lücken? (Tipp: Für Wetterdaten bietet sich Vorwärts-Auffüllen `ffill()` oder Interpolation an, kein Drop von ganzen Zeilen!).
- [ ] **Zeit-Features extrahieren (Zyklizität):**
    - Neue Spalten aus `Datum_Uhrzeit` generieren: `Monat`, `Wochentag`, `Stunde`.
    - *(Bonus für Fortgeschrittene: Sinus/Cosinus-Transformation für die Uhrzeit, damit das Modell versteht, dass 23 Uhr und 1 Uhr nah beieinander liegen).*
- [ ] **Lag-Features (Verzögerungen) berechnen:**
    - Erstellen von Spalten wie `Temperatur_vor_24h` oder `Ozon_gestern`. Das ist der wichtigste Prädiktor für Zeitreihen!
- [ ] **Rollierende Fenster (Moving Averages):**
    - Gleitender Durchschnitt der letzten 24h für Temperatur oder Windgeschwindigkeit berechnen.
- [ ] **Daten-Export:**
    - Den fertigen, bereinigten Datensatz als `train_ready_data.csv` im Unterordner speichern.
